```python
import pandas as pd 
import numpy as np
df=pd.read_csv('sample_titanic_data.csv')
df.head()
```

```python
df['number'].unique()
```

```python
#extract the numericl part from number column 
df['numeric_data']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
df['categorial_data']=np.where(df['numeric_data'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['ticket_numeric']=df['Ticket'].apply(lambda x:x.split()[-1])
df['ticket_numeric']=pd.to_numeric(df['ticket_numeric'],errors='coerce',downcast='integer')
df['ticket_cat']=df['Ticket'].apply(lambda x:x.split()[0])
df['ticket_cat']=np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
df.head()

```

```python
df['cabin_num']=df['Cabin'].str.extract(r'(\d+)')
df['cabin_cat']=df['Cabin'].str[0]
df.head()
```
