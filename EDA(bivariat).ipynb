```python
import pandas as pd
import seaborn as sns
titanic=pd.read_csv('tested.csv')
tips=sns.load_dataset('tips')
df=pd.DataFrame(tips)
titanic.head()
df.head()
```

```python
##scatterplot(n_n)bi
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
#multi
#sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=df['sex'])
#or
sns.scatterplot(data=tips,x="total_bill",y="tip",hue=df['sex'],style=df['smoker'],size=df['size'])
```

```python
#n-c
sns.barplot(x=tips['time'],y=tips['tip'],hue=tips['sex'])
```
